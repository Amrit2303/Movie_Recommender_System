# Content based movie recommender system

```python
# important columns to be kept
# genre, id, keywords, title,overview, cast,crew
```

```python
d_new=d_new[['genres','id','keywords','title','overview','cast','crew']]
```

```python
d_new.head()
```

```python
d_new.dropna(inplace=True)
```

```python
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert_use(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
d_new['genres']=d_new['genres'].apply(convert_use)
```

```python
d_new
```

```python
d_new['keywords']=d_new['keywords'].apply(convert_use)
```

```python
d_new.head()
```

```python
def convert_use_1(obj):
    L=[]
    count=1
    for i in ast.literal_eval(obj):
        if count!=3:
            L.append(i['name'])
            count+=1
        else:
            break
    return L
```

```python
d_new['cast'].apply(convert_use_1)
```

```python
d_new['cast']=d_new['cast'].apply(convert_use_1)
```

```python
d_new.head()
```

```python
def director_name(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
d_new['crew']=d_new['crew'].apply(director_name)
```

```python
d_new.head()
```

```python
d_new['overview'][0]
```

```python
d_new['overview']=d_new['overview'].apply(lambda x:x.split())
```

```python
d_new.head()
```

```python
def remove_spaces(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
d_new['cast'] = d_new['cast'].apply(remove_spaces)
d_new['crew'] = d_new['crew'].apply(remove_spaces)
d_new['genres'] = d_new['genres'].apply(remove_spaces)
d_new['keywords'] = d_new['keywords'].apply(remove_spaces)
```

```python
d_new['crew'][0]
```

```python
d_new['tags']=d_new['overview']+d_new['genres']+d_new['keywords']+d_new['crew']+d_new['cast']
```

```python
d_new.head()
```

```python
final_df=d_new[['id','title','tags']]
```

```python
final_df.head()
```

```python
final_df['tags']=final_df['tags'].apply(lambda x:" ".join(x))
```

```python
final_df.head()
```

```python
final_df['tags'][0]
```

```python
final_df['tags']=final_df['tags'].apply(lambda s:s.lower())
```

```python
final_df.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
def stem_Word(text):
    y=[]
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
stem_Word('In the 22nd century, a paraplegic Marine is dispatched to the moon Pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. Action Adventure Fantasy ScienceFiction cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d JamesCameron SamWorthington ZoeSaldana')
```

```python
final_df['tags']=final_df['tags'].apply(stem_Word)
```

```python
final_df.head()
```

# We'll use text to vector for model building purpose

```python
# bag of words
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector=cv.fit_transform(final_df['tags']).toarray()
```

```python
vector
```

```python
vector.shape
```

```python
cv.get_feature_names()
```

```python
len(cv.get_feature_names())
```

# We'll use cosine distance i.e. angle b/w 2 vectors as euclidean distance is not good for large dimensional space

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
similarity.shape
```

```python
similarity[0]
```

```python
list(enumerate(similarity[0]))
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])
```

```python
def recommend_movie(film):
    film_index=final_df[final_df['title']==film].index[0]
    distances=similarity[film_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(final_df.iloc[i[0]].title)
```

```python
final_df.iloc[2405].title
```

```python
recommend_movie('Batman Begins')
```

```python
recommend_movie('Avatar')
```
